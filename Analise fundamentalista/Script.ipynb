```python
import fundamentus
import numpy as np
import pandas as pd
```

```python
cot= fundamentus.get_resultado()
df = pd.DataFrame(cot)
```

```python
df.to_excel('./cotacao.xlsx' , index=True)
df = df[df['patrliq'] > 500000000]
```

```python
percentil_dy = np.percentile(df['dy'], 90)
```

```python
percentil_10_DY = df[df['dy'] > percentil_dy]
print(percentil_10_DY)

percentil_10_DY['dy'].to_excel('./ranking_dv.xlsx' , index=True)
```

```python
df_pl = df[df['pl'] > 0]

print(df_pl)
```

```python
percentil_pl = np.percentile(df_pl['pl'], 10)

print(percentil_pl)
```

```python
percentil_10_PL = df_pl[df_pl['pl'] < percentil_pl]
print(percentil_10_PL)

percentil_10_PL['pl'].to_excel('./ranking_pl.xlsx' , index=True)
```

```python
percentil_roe = np.percentile(df['roe'], 90)
percentil_10_ROE = df[df['roe'] > percentil_roe]
print(percentil_10_ROE)

percentil_10_ROE['roe'].to_excel('./ranking_roe.xlsx' , index=True)
```

```python
percentil_roic = np.percentile(df['roic'], 90)
percentil_10_ROIC = df[df['roic'] > percentil_roic]
print(percentil_10_ROIC)

percentil_10_ROIC['roic'].to_excel('./ranking_roic.xlsx' , index=True)
```

```python
percentil_10_PSR = df[(df['psr'] < 1) & (df['psr'] > 0)]
print(percentil_10_PSR)

percentil_10_PSR['psr'].to_excel('./ranking_psr.xlsx' , index=True)
```
