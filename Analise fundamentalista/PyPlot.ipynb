```python
# O código a seguir para criar um dataframe e remover as linhas duplicadas sempre é executado e age como um preâmbulo para o script: 

# dataset = pandas.DataFrame(dy)
# dataset = dataset.drop_duplicates()

# Cole ou digite aqui seu código de script:
import fundamentus
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


cot= fundamentus.get_resultado()
df = pd.DataFrame(cot)

percentil_dy = np.percentile(df['dy'], 90)

percentil_dy = df[df['dy'] > percentil_dy]
```

```python
percentil_dy = percentil_dy.sort_values(["dy"], ascending = [True])
```

```python
name=percentil_dy.index
price= percentil_dy['dy']

# Figure Size
fig, ax = plt.subplots(figsize =(20, 9))
 
# Horizontal Bar Plot
ax.barh(name, price)
 
# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)
 
# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
 
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
 
# Add x, y gridlines
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
 
# Show top values
ax.invert_yaxis()
 
# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')
 
# Add Plot Title
ax.set_title('Ranking por DY',
             loc ='left', )
 
# Add Text watermark
fig.text(0.9, 0.15, 'Thiago Ezaki', fontsize = 12,
         color ='grey', ha ='right', va ='bottom',
         alpha = 0.7)
 
# Show Plot
plt.savefig(".\contratos.png" , dpi=600, bbox_inches='tight')
```
